# Cross-Sectional Factor Model (per time t)

**Model**
$$
r_{it}=\alpha_t+\sum_{j=1}^K X_{ij,t}\,\beta_{jt}+\varepsilon_{it},
\qquad
\mathbf{r}_t=\mathbf{1}\alpha_t+\mathbf{X}_t\,\boldsymbol{\beta}_t+\boldsymbol{\varepsilon}_t.
$$

**Estimation (OLS / WLS)**  
Let $\mathbf{Z}_t=[\mathbf{1}\ \ \mathbf{X}_t]$, $\boldsymbol{\theta}_t=[\alpha_t;\ \boldsymbol{\beta}_t]$.

OLS:
$$
\hat{\boldsymbol{\theta}}_t=(\mathbf{Z}_t^\top\mathbf{Z}_t)^{-1}\mathbf{Z}_t^\top\mathbf{r}_t.
$$

WLS (optional), with $\boldsymbol{\Omega}_t=\mathrm{diag}(\omega_{i,t}\ge 0)$:
$$
\hat{\boldsymbol{\theta}}_t=(\mathbf{Z}_t^\top\boldsymbol{\Omega}_t\mathbf{Z}_t)^{-1}\mathbf{Z}_t^\top\boldsymbol{\Omega}_t\mathbf{r}_t.
$$

Residuals:
$$
\hat{\boldsymbol{\varepsilon}}_t=\mathbf{r}_t-\mathbf{1}\hat{\alpha}_t-\mathbf{X}_t\hat{\boldsymbol{\beta}}_t.
$$

**Factor time-series objects**
$$
\hat{\boldsymbol{\mu}}_\beta=\frac{1}{T}\sum_{t=1}^T \hat{\boldsymbol{\beta}}_t,\qquad
\hat{\boldsymbol{\Sigma}}_\beta=\frac{1}{T-1}\sum_{t=1}^T(\hat{\boldsymbol{\beta}}_t-\hat{\boldsymbol{\mu}}_\beta)(\hat{\boldsymbol{\beta}}_t-\hat{\boldsymbol{\mu}}_\beta)^\top.
$$

# Portfolio Factor Analytics

Original (signed per-NAV) weights $\mathbf{w}_{0,t}$ with $\mathbf{1}^\top\mathbf{w}_{0,t}=1$.

Exposures:
$$
\mathbf{b}_{0,t}=\mathbf{X}_t^\top \mathbf{w}_{0,t}.
$$

Return decomposition:
$$
r_{p,t}=\hat{\alpha}_t+\mathbf{b}_{0,t}^\top \hat{\boldsymbol{\beta}}_t+\mathbf{w}_{0,t}^\top \hat{\boldsymbol{\varepsilon}}_t.
$$

Risk:
$$
\operatorname{Var}(r_{p})=\mathbf{b}_{0,t}^\top \hat{\boldsymbol{\Sigma}}_\beta\,\mathbf{b}_{0,t}
+\mathbf{w}_{0,t}^\top \hat{\boldsymbol{\Sigma}}_{\varepsilon,t}\,\mathbf{w}_{0,t}.
$$

---

# Hedging to Target Exposures

Target exposures $\mathbf{b}_t^*\in\mathbb{R}^K$.  
Hedge weights $\mathbf{w}_{h,t}$ (budget-neutral or not) with:
$$
\mathbf{X}_t^\top (\mathbf{w}_{0,t}+\mathbf{w}_{h,t})=\mathbf{b}_t^*,
\qquad
\Delta\mathbf{b}_t:=\mathbf{b}_t^*-\mathbf{b}_{0,t}=\mathbf{X}_t^\top \mathbf{w}_{h,t}.
$$

---

## Full-Portfolio Quadratic Program (budgeted)

Choose final $\mathbf{w}_t$:

**Objective**
$$
\min_{\mathbf{w}_t}\ \frac12(\mathbf{X}_t^\top\mathbf{w}_t-\mathbf{b}_t^*)^\top
\mathbf{W}_t(\mathbf{X}_t^\top\mathbf{w}_t-\mathbf{b}_t^*)
+\frac{\lambda}{2}\,\mathbf{w}_t^\top \mathbf{C}_t \mathbf{w}_t
\quad\text{s.t. }\mathbf{1}^\top\mathbf{w}_t=1,
$$
with $\mathbf{W}_t\succeq 0$ (factor metric; e.g., $\mathbf{I}_K$ or $\hat{\boldsymbol{\Sigma}}_\beta$),
$\mathbf{C}_t\succeq 0$ (asset penalty; e.g., $\hat{\boldsymbol{\Sigma}}_{\varepsilon,t}$), $\lambda\ge 0$.

**KKT / Closed form**  
Let
$$
\mathbf{A}_t=\mathbf{X}_t\mathbf{W}_t\mathbf{X}_t^\top+\lambda \mathbf{C}_t,\qquad
\mathbf{c}_t=\mathbf{X}_t\mathbf{W}_t\mathbf{b}_t^*.
$$
Then
$$
\mathbf{A}_t\mathbf{w}_t=\mathbf{c}_t+\mu_t\mathbf{1},\qquad
\mu_t=\frac{1-\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{c}_t}{\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{1}},
$$
$$
\boxed{
\mathbf{w}_t=\mathbf{A}_t^{-1}\mathbf{c}_t
+\frac{\,1-\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{c}_t\,}{\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{1}}\,
\mathbf{A}_t^{-1}\mathbf{1}.
}
$$

---

## Hedge-Only (budget-neutral)

Solve for $\mathbf{w}_{h,t}$ with $\mathbf{1}^\top\mathbf{w}_{h,t}=0$ and target change $\Delta \mathbf{b}_t$:

$$
\min_{\mathbf{w}_{h,t}}\ \frac12(\mathbf{X}_t^\top\mathbf{w}_{h,t}-\Delta\mathbf{b}_t)^\top
\mathbf{W}_t(\mathbf{X}_t^\top\mathbf{w}_{h,t}-\Delta\mathbf{b}_t)
+\frac{\lambda}{2}\,\mathbf{w}_{h,t}^\top \mathbf{C}_t \mathbf{w}_{h,t}
\quad\text{s.t. }\mathbf{1}^\top\mathbf{w}_{h,t}=0.
$$

Closed form (same $\mathbf{A}_t$):
$$
\mu_t=-\frac{\mathbf{1}^\top \mathbf{A}_t^{-1}(\mathbf{X}_t\mathbf{W}_t\Delta\mathbf{b}_t)}{\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{1}},
\qquad
\boxed{
\mathbf{w}_{h,t}=\mathbf{A}_t^{-1}(\mathbf{X}_t\mathbf{W}_t\Delta\mathbf{b}_t)
-\frac{\mathbf{1}^\top \mathbf{A}_t^{-1}(\mathbf{X}_t\mathbf{W}_t\Delta\mathbf{b}_t)}{\mathbf{1}^\top \mathbf{A}_t^{-1}\mathbf{1}}\,
\mathbf{A}_t^{-1}\mathbf{1}.
}
$$

Final weights: $\mathbf{w}_{\text{final},t}=\mathbf{w}_{0,t}+\mathbf{w}_{h,t}$.

---

## Exact Exposure Matching (optional, QP)

If you require $\,\mathbf{X}_t^\top\mathbf{w}_t=\mathbf{b}_t^*\,$ exactly (min specific risk):
$$
\min_{\mathbf{w}_t}\ \tfrac12\,\mathbf{w}_t^\top \mathbf{C}_t \mathbf{w}_t
\quad\text{s.t.}\quad
\mathbf{X}_t^\top \mathbf{w}_t=\mathbf{b}_t^*,\ \ \mathbf{1}^\top\mathbf{w}_t=1.
$$


# Target Exposure Formulations — \( \mathbf{b}_t^* \)

## 1. Full Neutralization (Alpha-only)

Set selected factor exposures to zero:
$$
(\mathbf{b}_t^*)_k =
\begin{cases}
0, & k \in \mathcal{K}, \\
\text{free or benchmark}, & k \notin \mathcal{K}.
\end{cases}
$$

Emphasize these factors in the weight matrix \( \mathbf{W}_t \) to enforce stronger hedging.

---

## 2. Benchmark-relative Target

Benchmark exposure:
$$
\mathbf{b}_{\text{bench},t} = \mathbf{X}_t^\top \mathbf{w}_{\text{bench},t}.
$$

Target exposure:
$$
\boxed{\mathbf{b}_t^* = \mathbf{b}_{\text{bench},t} + \Delta \mathbf{b}_t},
\qquad
\underline{\Delta} \le \Delta \mathbf{b}_t \le \overline{\Delta}.
$$

---

## 3. Mean–Variance Optimal in Factor Space

Maximize factor-level Sharpe ratio:
$$
\max_{\mathbf{b}}\;
\mathbf{b}^\top \boldsymbol{\mu}_{\beta,t}
- \frac{\gamma}{2}\mathbf{b}^\top \boldsymbol{\Sigma}_{\beta,t}\mathbf{b}.
$$

Closed form:
$$
\boxed{
\mathbf{b}_t^* = \frac{1}{\gamma}\,
\boldsymbol{\Sigma}_{\beta,t}^{-1}\boldsymbol{\mu}_{\beta,t}.
}
$$

### 3a. With Factor Risk Budget \( R^* \)

Normalize to target total factor variance:
$$
\tilde{\mathbf{b}} = \boldsymbol{\Sigma}_{\beta,t}^{-1}\boldsymbol{\mu}_{\beta,t},
\qquad
\boxed{
\mathbf{b}_t^* =
\sqrt{\frac{R^*}{\tilde{\mathbf{b}}^\top \boldsymbol{\Sigma}_{\beta,t}\tilde{\mathbf{b}}}}\;
\tilde{\mathbf{b}}.
}
$$

### 3b. With Linear Constraints

$$
\begin{aligned}
\max_{\mathbf{b}}\;&
\mathbf{b}^\top \boldsymbol{\mu}_{\beta,t}
- \frac{\gamma}{2}\mathbf{b}^\top \boldsymbol{\Sigma}_{\beta,t}\mathbf{b} \\
\text{s.t. }&
A\mathbf{b}=c, \quad
\underline{\mathbf{b}} \le \mathbf{b} \le \overline{\mathbf{b}}.
\end{aligned}
$$

---

## 4. Equal Risk Contribution (ERC)

Each factor contributes equally to total variance.

Risk contribution:
$$
\mathrm{RC}_k(\mathbf{b}) = b_k (\boldsymbol{\Sigma}_{\beta,t}\mathbf{b})_k.
$$

ERC condition:
$$
\boxed{
\mathrm{RC}_k(\mathbf{b}_t^*) = \rho_k\,V(\mathbf{b}_t^*),
\qquad
V(\mathbf{b}) = \mathbf{b}^\top \boldsymbol{\Sigma}_{\beta,t}\mathbf{b}.
}
$$

---

## 5. Scenario / Robust Target

For stress-testing scenarios \( s = 1,\ldots,S \):
$$
\boxed{
\mathbf{b}_t^* =
\arg\min_{\mathbf{b}}\;
\max_s
\big[
\mathbf{b}^\top \boldsymbol{\Sigma}_{\beta,t}^{(s)} \mathbf{b}
- 2\,c^{(s)\top}\mathbf{b}
\big]
\quad
\text{s.t. }
\underline{\mathbf{b}} \le \mathbf{b} \le \overline{\mathbf{b}}.
}
$$

---

## 6. Black–Litterman Factor Target

Posterior expected factor premia:
$$
\boldsymbol{\mu}_{\beta,t}^{\text{BL}}
=
\Big(
\frac{1}{\tau}\boldsymbol{\Sigma}_{\beta,t}^{-1}
+ P^\top \boldsymbol{\Omega}^{-1} P
\Big)^{-1}
\Big(
\frac{1}{\tau}\boldsymbol{\Sigma}_{\beta,t}^{-1}\boldsymbol{\mu}_0
+ P^\top \boldsymbol{\Omega}^{-1}\mathbf{q}
\Big).
$$

Target exposure:
$$
\boxed{
\mathbf{b}_t^* = \frac{1}{\gamma}\,
\boldsymbol{\Sigma}_{\beta,t}^{-1}\,
\boldsymbol{\mu}_{\beta,t}^{\text{BL}}.
}
$$

---

## 7. Shrink-to-Anchor (Turnover Control)

Blend current and anchor exposures:
$$
\boxed{
\mathbf{b}_t^* = (1-\alpha)\,\mathbf{b}_{0,t}
+ \alpha\,\mathbf{b}_t^{\text{anchor}},
\qquad
\alpha \in [0,1].
}
$$

---

## 8. Norm and Leverage Controls

$$
\mathbf{b}_t^{*\top} \boldsymbol{\Sigma}_{\beta,t}\mathbf{b}_t^* \le R^*, 
\qquad
\underline{\mathbf{b}} \le \mathbf{b}_t^* \le \overline{\mathbf{b}}.
$$

---

## 9. Plug-in for Hedge Optimization

After determining \( \mathbf{b}_t^* \), insert it into one of the hedge optimization problems:

**Full-portfolio (budgeted):**
$$
\min_{\mathbf{w}_t}
\frac{1}{2}(\mathbf{X}_t^\top\mathbf{w}_t-\mathbf{b}_t^*)^\top
\mathbf{W}_t(\mathbf{X}_t^\top\mathbf{w}_t-\mathbf{b}_t^*)
+\frac{\lambda}{2}\mathbf{w}_t^\top \mathbf{C}_t \mathbf{w}_t,
\qquad
\mathbf{1}^\top\mathbf{w}_t=1.
$$

**Hedge-only (budget-neutral):**
$$
\min_{\mathbf{w}_{h,t}}
\frac{1}{2}(\mathbf{X}_t^\top\mathbf{w}_{h,t}-\Delta\mathbf{b}_t)^\top
\mathbf{W}_t(\mathbf{X}_t^\top\mathbf{w}_{h,t}-\Delta\mathbf{b}_t)
+\frac{\lambda}{2}\mathbf{w}_{h,t}^\top \mathbf{C}_t \mathbf{w}_{h,t},
\qquad
\mathbf{1}^\top\mathbf{w}_{h,t}=0,
$$

where \( \Delta\mathbf{b}_t = \mathbf{b}_t^* - \mathbf{b}_{0,t} \).
